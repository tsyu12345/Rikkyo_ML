## 宿題1



```python
import math

#ユークリッド距離を求める関数
def euclid(data1: list[int], data2: list[int]) -> float:
    #2データの次元数が異なる場合はエラーを返す
    if len(data1) != len(data2):
        raise ValueError("2データの次元数が異なります, data1: {}, data2: {}".format(len(data1), len(data2)))
    
    #次元数n
    n: int = len(data1)
    #ユークリッド距離を計算し返す
    return math.sqrt(sum([(data1[i] - data2[i]) ** 2 for i in range(n)]))


#ユークリッド距離を計算し、距離が近いデータペア順に出力
data: list[list[int]] = [
    [5,2,5,8],
    [7,10,1,12],
    [3,2,6,3]
]

#結果を格納するリスト
result: list[tuple[float, list[int], list[int]]] = []

for v in data:
    for w in data:
        if v != w:
            #ユークリッド距離を計算
            d: float = euclid(v, w)
            #計算結果をリストに格納
            result.append((d, v, w))
#距離が近い順にソート
result.sort()
print(*result, sep="\n")
```

## 宿題2





```python
#コサイン類似度を求める関数
def cos_sim(data1: list[int], data2: list[int]) -> float:
    #2データの次元数が異なる場合はエラーを返す
    if len(data1) != len(data2):
        raise ValueError("2データの次元数が異なります, data1: {}, data2: {}".format(len(data1), len(data2)))
    
    #次元数n
    n: int = len(data1)
    #内積を計算
    inner_product: float = sum([data1[i] * data2[i] for i in range(n)])
    #ノルムを計算
    norm1: float = math.sqrt(sum([data1[i] ** 2 for i in range(n)]))
    norm2: float = math.sqrt(sum([data2[i] ** 2 for i in range(n)]))
    #コサイン類似度を計算し返す
    return inner_product / (norm1 * norm2)

data: list[list[int]] = [
    [3,2,1],
    [10,6,3],
    [-1,3,2]
]

result: list[tuple[float, list[int], list[int]]] = []

for v in data:
    for w in data:
        if v != w:
            #コサイン類似度を計算
            d: float = cos_sim(v, w)
            #計算結果をリストに格納
            result.append((d, v, w))
#類似度が高い順にソート cos類似度は大きいほど、データ間距離が近い
result.sort(reverse=True)

print(*result, sep="\n")
```
